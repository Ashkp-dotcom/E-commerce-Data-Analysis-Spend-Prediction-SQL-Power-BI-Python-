# Machine Learning - Spending Prediction 

```python
# Spending prediction per user - Supervised Approach / Linear regression 
```

## One hot encoding

```python
user_agg['top_channel'] = user_agg['top_channel'].astype(str)
user_agg['top_state'] = user_agg['top_state'].astype(str)
```

```python
encoded_df = pd.get_dummies(user_agg, columns=['top_channel', 'top_state'], drop_first=True)
```

```python
encoded_df
```

```python
encoded_df = encoded_df.astype({col: int for col in encoded_df.select_dtypes(bool).columns})
```

```python
encoded_df
```

```python
X = encoded_df.drop('total_spent', axis=1)
y = encoded_df['total_spent']
```

```python
X.head()
```

```python
y.head()
```

## Train Test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=78)
```

## Standardisation 

```python
# Fit and transform training data

ss=StandardScaler()

X_train_ft = ss.fit_transform(X_train)
X_train_ft = pd.DataFrame(X_train_ft, columns=X_train.columns)

# Transform test data using the same scale
X_test_ft = ss.transform(X_test)
X_test_ft = pd.DataFrame(X_test_ft, columns=X_test.columns)
```

## Model Training 

### Linear Regression 

```python
model = LinearRegression()
model.fit(X_train_ft, y_train)
```

## Prediction 

```python
y_pred = model.predict(X_test_ft)
```

## Model Evaluation 

```python
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.4f}")
```

### Random forest 

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train_ft, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test_ft)
```

```python
rmse_rf = mean_squared_error(y_test, y_pred_rf, squared=False)
r2_rf = r2_score(y_test, y_pred_rf)
```

```python
print(f"Random Forest RMSE: {rmse_rf:.2f}")
print(f"Random Forest R² Score: {r2_rf:.4f}")
```
